```python
import numpy as np
import pandas as pd


path = '/opt/ml/input/data/'

train = pd.read_csv(path + 'train_data.csv')
test = pd.read_csv(path + 'test_data.csv')
```

```python
train
```

```python
time = pd.to_datetime(train['Timestamp'])
df = pd.DataFrame()
df['Timestamp'] = time
df['Timestamp_quarter']       = df['Timestamp'].dt.quarter       # 분기(숫자)
df['Timestamp_quarter'].value_counts()
```

```python
df['Timestamp_weekday']       = df['Timestamp'].dt.weekday       # 요일숫자(0-월, 1-화) (=dayofweek)
df['Timestamp_weekday'].value_counts()
```

```python
df['Timestamp_dayofyear']     = df['Timestamp'].dt.dayofyear     # )연 기준 몇일째(숫자
df['Timestamp_dayofyear'].value_counts()
```

```python
#df['Timestamp_days_in_month'] = df['Timestamp'].dt.days_in_month # 월 일수(숫자) (=daysinmonth)
#df['Timestamp_days_in_month'].value_counts()
```

```python
df['Timestamp_week']    = df['Timestamp'].dt.week    # 연 기준 몇주째(숫자) (=week)
df['Timestamp_week'].value_counts()
```
